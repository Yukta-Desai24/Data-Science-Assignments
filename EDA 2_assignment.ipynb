# Scaling Techniques:

**Standard Scaling (Z-score normalization):**
It scales the data so that it has a mean of 0 and a standard deviation of 1.

Formula:     Z = (x−μ) / σ

Usage:

1) Preferred when the distribution of data is Gaussian (normal distribution).

2) Useful for algorithms that assume normally distributed data, such as linear regression, logistic regression, and SVMs.

3) Helps in cases where the standard deviation is significant or where outliers are present, as it minimizes the impact of outliers in the scaling process.

**Min-Max Scaling:**
It scales the data to a fixed range, typically between 0 and 1.

Formula:     X(scaled) = {x – min(x)} / {max(x) – min(x)}

Usage:

1) Suitable when the distribution of data is not Gaussian and the algorithm does not assume any specific distribution of the data.

2) Neural networks (especially those using gradient descent optimization) often perform better with inputs scaled to a smaller range like [0, 1] or [-1, 1].

3) Helps in cases where there are small standard deviations or when the scale of the data varies widely.

**Conclusion:**

• Standard Scaling: Use when the data follows a normal distribution or when working with algorithms that assume normally distributed data. It is also useful when you need to compare different features that have different units of measurement.

• Min-Max Scaling: Use when the distribution of your data does not follow a normal distribution or when you know the distribution is bounded (like percentages between 0 and 100). It is particularly useful for neural networks, image processing, and clustering algorithms like K-Means.

# Encoding Techniques:

•	Apply One-Hot Encoding to categorical variables with less than 5 categories.

**sex, race, income**

•	Use Label Encoding for categorical variables with more than 5 categories.

**workclass, education, marital_status, occupation, relationship**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse=False)
# Initialize LabelEncoder
label_encoder = LabelEncoder()
```

```python
# Encoding gender column:

encoded_gender = one_hot_encoder.fit_transform(data[['sex']])
encoded_feature_names = one_hot_encoder.get_feature_names_out(['sex'])

# Create a DataFrame with the one-hot encoded data and appropriate column names
one_hot_encoded_df = pd.DataFrame(encoded_gender , columns=encoded_feature_names)

data = pd.concat([data, one_hot_encoded_df], axis=1)
```

```python
# Encoding race column:

encoded_race = one_hot_encoder.fit_transform(data[['race']])
encoded_feature_race = one_hot_encoder.get_feature_names_out(['race'])

# Create a DataFrame with the one-hot encoded data and appropriate column names
one_hot_encoded_df = pd.DataFrame(encoded_race , columns=encoded_feature_race)

data = pd.concat([data, one_hot_encoded_df], axis=1)
```

```python
# Encoding income column:

encoded_income = one_hot_encoder.fit_transform(data[['income']])
encoded_feature_income = one_hot_encoder.get_feature_names_out(['income'])

# Create a DataFrame with the one-hot encoded data and appropriate column names
one_hot_encoded_df = pd.DataFrame(encoded_income , columns=encoded_feature_income)

data = pd.concat([data, one_hot_encoded_df], axis=1)
```

```python
data['workclass'] = label_encoder.fit_transform(data['workclass'])
data['education'] = label_encoder.fit_transform(data['education'])
data['marital_Status'] = label_encoder.fit_transform(data['marital_status'])
data['occupation'] = label_encoder.fit_transform(data['occupation'])
data['relationship'] = label_encoder.fit_transform(data['relationship'])
```

**PROS AND CONS**

**One-Hot Encoding**

**Pros:**

1) No Ordinal Relationship: Suitable for categorical variables without any intrinsic order.

2) Avoids Ordinal Issues: Prevents the model from assuming any ordinal relationship between categories.

3) Widely Used: Commonly used in many machine learning algorithms and frameworks.

**Cons:**

1) High Dimensionality: Can significantly increase the dimensionality of the dataset, especially with many unique categories.

2) Sparse Data: Results in a sparse matrix, which can be memory inefficient.

3) Scalability: May not scale well with large datasets or categories.

**Label Encoding**

**Pros:**

1) Simplicity: Easy to implement and understand.

2) Low Dimensionality: Does not increase the number of features in the dataset.

3) Memory Efficient: Suitable for algorithms that do not require additional dimensions for categorical features.

**Cons:**

1) Ordinal Relationship Assumption: Implies an ordinal relationship between categories, which can mislead some machine learning models.

2) Less Suitable for Nominal Data: Not suitable for nominal categorical variables without an inherent order.

3) Model Bias: Can introduce bias into the model if the numerical order is not meaningful.

# Feature Engineering:

**Categorizing age into different groups - Young Adult, Middle Age, Senior:**

```python
bins = [0, 35, 55, float('inf')]
labels = ['Young Adult', 'Middle Age', 'Senior']
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)
```

```python
data.head()
```

**Age** can be a significant factor in determining income levels and occupation choices. Grouping ages into categories can help capture non-linear relationships between age and income.

Different **age groups** may have different likelihoods of earning higher incomes or being in different types of occupations.

**Categorizing education level into different groups - High School, College, Degree/Higher:**

```python
bins = [0, 8, 12, 16]
labels = ['High School', 'College', 'Degree or Higher']
data['edu_level'] = pd.cut(data['education_num'], bins=bins, labels=labels, right=False)
```

By categorizing the **education level** we can determine based on these categorization does the workclass of the people affect and thus accodingly does there income level depend on the education level.

**Applying log transformation on the numerical column capital_gain:**

```python
# Apply log transformation to 'capital_gain'
data['capital_gain_log'] = data['capital_gain'].apply(lambda x: np.log(x + 1))

print(data[['capital_gain', 'capital_gain_log']])
```

The column capital_gain has a **positively skewed distribution.**

The log transformation is chosen **to reduce skewness and stabilize the variance** of the positively skewed 'capital_gain' variable, making the data more normally distributed.

**PPS Matrix:** The PPS matrix provides a score between 0 and 1 for each pair of features, indicating the strength of the predictive power of one feature for another. A PPS of 0 means no predictive power, while a PPS of 1 means perfect predictive power.

**Correlation Matrix:** The correlation matrix provides a Pearson correlation coefficient for each pair of features, indicating the strength and direction of a linear relationship between features. The coefficient ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation).
