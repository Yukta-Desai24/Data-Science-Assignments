**Handling missing values:**

```python
data.isnull().sum()
```

```python
missing_values = ['LB', 'AC', 'DS', 'DP', 'MLTV', 'Width', 'Tendency', 'NSP']
```

```python
for col in missing_values:
    sns.displot(data[col], kde=True, palette="viridis")
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()
```

**Variables LB and Width** is approximately normally distributed. Hence using **mean** to replace the missing values.

Whereas **other variables** are positively skewed. Hence using **median** value as it is robust to outliers.

For **Tendency and NSP** using **mode** values.

```python
for col in ['LB', 'Width']:
    data[col].fillna(data[col].mean(), inplace=True)
```

```python
for col in ['AC', 'DS', 'DP', 'MLTV']:
    data[col].fillna(data[col].median(), inplace=True)
```

```python
for col in ['NSP', 'Tendency']:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
data.isnull().sum()
```

**Now the data has no missing values.**

```python
# Convert 'NSP' and 'Tendency' columns to integers
data['NSP'] = data['NSP'].astype(int)
data['Tendency'] = data['Tendency'].astype(int)
```

**Outlier analysis:**

```python
# Dividing the variables as categorical and numerical variables:-
cat_list=[]
num_list=[]


for i in data.columns:
    unique_values = len(data[i].unique())
    if unique_values<10:
        cat_list.append(i)
    else:
        num_list.append(i)
```

```python
print(cat_list)
print('\n',num_list)
```

```python
k=1
plt.figure(figsize=(13,13))
plt.suptitle("Distribution of Outliers")

for i in data.loc[:,num_list]:
    plt.subplot(5,3,k)
    sns.boxplot(x = i, data = data.loc[:,num_list], palette = 'rocket_r')
    plt.title(i)
    plt.tight_layout()
    k+=1
```

```python
out_list = ['LB', 'AC', 'UC', 'DL', 'ASTV', 'MSTV', 'Width']
```

```python
# remove outliers;
for i in data.loc[:,out_list]:
    
    Q1 = data[i].quantile(0.25)
    Q3 = data[i].quantile(0.75)
    IQR = Q3 - Q1
    up_lim=Q3+1.5 *IQR
    low_lim=Q1-1.5 *IQR
    
    data.loc[data[i]>up_lim,i]=up_lim
    data.loc[data[i]<low_lim,i]=low_lim
```

```python
k=1
plt.figure(figsize=(13,13))
plt.suptitle("Distribution of Outliers")

for i in data.loc[:,num_list]:
    plt.subplot(5,3,k)
    sns.boxplot(x = i, data = data.loc[:,num_list], palette = 'viridis')
    plt.title(i)
    plt.tight_layout()
    k+=1
```
